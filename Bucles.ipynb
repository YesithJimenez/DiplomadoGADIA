```python
#Bucle for El bucle for se utiliza para iterar sobre una secuencia (como una lista, tupla, diccionario, conjunto o cadena de caracteres). Se ejecuta una vez por cada elemento en la secuencia.

```

```python
# imprimer todos los numeros de una secuencia de datos

secuencia= [1, 2, 3, 4, 5]

for i in secuencia:
    print(i)
```

```python
#Calsifique los numeros que son pares y los que son impares de la lista secuencua

for i in secuencia:
    if i % 2 ==0:
      print(f"Elnumero{i} es par")
    else:
      print(f"Elnumero{i} es inpar")
```

```python
# contruir un contador que vaya hasta 5 y cuando la variable tenga este valor, deje de contar
contador=0
while contador<5:
    print(contador)
    contador+=1
```

#break: Sale del bucle antes de que se evalúe la condición o se termine de iterar sobre la secuencia.
continue: Salta el resto del código dentro del bucle y vuelve a la condición o a la siguiente iteración.
else: Se ejecuta una vez cuando el bucle termina, a menos que se haya interrumpido con break.

```python
Contraceña_Correcta="Segura123"
intentos=3
while intentos >0:
      contraceña = input("Introduce tu contraceña")
      if contraceña == Contraceña_Correcta:
        print("acceso concedido")
        break
      else:
        intentos-=1
        print(f"Contraceña incorrecta, te quedan {intentos} intentos")
if intentos == 0:
    print("Ya no te quedan intentos")
```

```python
import random

def adivinar_numero_secreto():
    numero_secreto = random.randint(1, 10)
    adivinanza = None

    print("Adivina el número secreto entre 1 y 10.")

    while adivinanza != numero_secreto:
        try:
            adivinanza = int(input("Ingresa tu número: "))

            if adivinanza < 1 or adivinanza > 10:
                print("Por favor, ingresa un número entre 1 y 10.")
            elif adivinanza < numero_secreto:
                print("El número secreto es mayor. Intenta de nuevo.")
            elif adivinanza > numero_secreto:
                print("El número secreto es menor. Intenta de nuevo.")
            else:
                print(f"¡Felicidades! Adivinaste el número secreto: {numero_secreto}")
        except ValueError:
            print("Por favor, ingresa un número válido.")

if __name__ == "__main__":
    adivinar_numero_secreto()
```
